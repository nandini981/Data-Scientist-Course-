```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
Rohit_Sharma = [200, 340, 200, 200, 2500, 1000, 1200, 1300, 1900, 822]

plt.plot(years, kohli,color='Green',linestyle="--", label="Virat Kohli")
plt.plot(years, sehwag,color='Red', linestyle="--", label="Virender Sehwag")
plt.plot(years, Rohit_Sharma,color='Blue', linestyle="-.",  label="Rohit_Sharma")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.style.use("ggplot")
plt.show()
```
