```python
import json
```

```python
def clean_data(data):
    #remove nulls, duplicates, and standardize formats
    data["users"] = [user for user in data["users"] if user["name"].strip()] # Remove users with empty names
    
    #remove duplicate friends lists
    for user in data["users"]:
        user["friends"] = list(set(user["friends"]))
    

    #remove inactive users
    data["users"] = [user for user in data["users"] if user["friends"] or user["liked_pages"]]


    # remove duplicate pages based on 'id'
    unique_pages = {}
    for page in data["pages"]:
        unique_pages[page["id"]] = page

    data["pages"] = list(unique_pages.values())
    return data



data = json.load(open('cleaning_data.json'))
data = clean_data(data)
json.dump(data, open('cleaned_data.json', 'w'), indent=4)
print("Data cleaning complete. Cleaned data saved to 'cleaned_data.json'.")
```
