```python
import json
```

```python
def load_data(filename):
    with open(filename, 'r') as file:
        data = json.load(file)
    return data
```

```python
data = load_data('data.json')
```

```python
#write a function to display users and their connections
def display_users_and_friends(data):
    for user in data['users']:
        name = user['name']
        friends = user['friends']
        id = user['id']
        print(f"ID: {id} , User: {name}, Friends: {friends}") 
    for page in data['pages']:
        id = page['id']
        book_name = page['name']
        print(f"id: {id} , Likes: {book_name}")
```

```python
display_users_and_friends(data)
```
