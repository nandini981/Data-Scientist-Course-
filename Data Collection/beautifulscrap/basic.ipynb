```python
from bs4 import BeautifulSoup
```

```python
with open("page1.html", "r") as f:
    content = f.read()
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
h3s = soup.find_all("h3")
```

```python
h3s
```

```python
for h3 in h3s:
    print(h3.a["title"])
```

```python
articles = soup.find_all("article", class_="product_pod")
```

```python
articles
```

```python
articles = soup.select("article.product_pod")
```

```python
# make sure `articles` is still a list of Tag objects
articles = soup.select("article.product_pod")
items =[]
for article in articles:
    title = article.find("h3").find("a")["title"]
    # correct method name and avoid running on an int
    price = article.select_one("p.price_color").text
    rating_element = article.select_one("p.star-rating")
    rating_classes = rating_element["class"]
    rating = [cls for cls in rating_classes if cls != "star-rating"][0]
    items.append((title, price, rating))
    print(title, price, rating)
   
```

```python
items
```

```python
import pandas as pd
```

```python
pd.DataFrame(items, columns=["Title", "Price", "Rating"])
```

```python
# create the dataframe from the parsed items (pandas is already imported
# in cell 11) and save it
df = pd.DataFrame(items, columns=["Title", "Price", "Rating"])
df.to_csv("books.csv", index=False)
```
