```python
import requests
```

```python
a = requests.get("https://books.toscrape.com/")
```

```python
for i in range(1, 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"page{i}.html", "w") as f:
        f.write(a.text)
        print(f"page{i}.html saved")
```
