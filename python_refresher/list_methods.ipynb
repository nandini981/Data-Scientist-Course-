```python
fruits = ["apple", "cherry", "banana"]
fruits.append("orange")
print(fruits)  # ['apple', 'cherry', 'banana', 'orange']

fruits.sort()
print(fruits)  # ['apple', 'banana', 'cherry', 'orange']
```

```python
lst1 = [1, 2, 3]
lst2 = lst1.copy()
lst2.append(4)
print("lst1:", lst1)  # lst1: [1, 2, 3]
print("lst2:", lst2)  # lst2: [1, 2, 3, 4]
print(lst1 + lst2)  # [1, 2, 3, 1, 2, 3, 4])
lst3 = lst1 + lst2
print(lst3)  # [1, 2, 3, 1, 2, 3, 4]
```

```python
lst3.sort()
```
