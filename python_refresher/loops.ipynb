
# Loops in Python


```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
for i in range(3):
    print(i)  # Output: 0, 1, 2
```

```python
count = 0
while count < 3:
    print(count)
    count += 1
```

```python
for count in range(1,11):
    print(count)
```

```python
for i in range(5):
    if i == 3:
        break  # Stops the loop at 3
    print(i)
```

```python
lst = [1, 2, 3, 4, 5]
for num in lst:
    print(num)
    if num == 3:
        break
    
```

```python
lst = [1, 2, 3, 4, 5]
for num in lst:
    if num == 3:
        continue
    print(num)
    
```
