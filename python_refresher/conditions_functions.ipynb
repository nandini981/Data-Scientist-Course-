## If else Conditionals & Functions in Python

In Python, conditional statements (if, elif, and else) are used to control the flow of a program based on conditions. These are essential in data science for handling different scenarios in data processing, decision-making, and logic execution.



```python
age = int(input("Enter your age: "))
print("You are " + str(age) + " years old.")
```

```python
if age < 18:
    print("You are a minor.")
else:
    print("You are an adult.")
```

```python
score = float(input("Enter your score: "))
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
age = 25
if age < 18:
    category = "Minor"
elif age < 65:
    category = "Adult"
else:
    category = "Senior Citizen"

print("Category:", category)
```

## functions


```python
def avg(a, b):
    return (a + b) / 2
```

```python
avg_score = avg(85, 95)
print("Average Score:", avg_score)
```
